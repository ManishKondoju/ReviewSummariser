```python
from datasets import Dataset

# Load using pandas
df_new = pd.read_csv("ReviewLarge.csv")

# Convert to Hugging Face dataset
dataset = Dataset.from_pandas(df_new)
```

```python
dataset = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = dataset["train"]
test_dataset = dataset["test"]
```

# Preprocessing Dataset

```python
def preprocess_function(example):
    return {
        "input_text": "summarize: " + example["review"],
        "target_text": example["summary"]
    }

train_data = train_dataset.map(preprocess_function)
test_data = test_dataset.map(preprocess_function)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")

def tokenize_function(examples):
    model_inputs = tokenizer(
        examples["input_text"], max_length=512, truncation=True, padding="max_length"
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["target_text"], max_length=64, truncation=True, padding="max_length"
        )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_train = train_data.map(tokenize_function, batched=True)
tokenized_test = test_data.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForSeq2SeqLM, TrainingArguments, Trainer

model_config1 = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

training_args_config1 = TrainingArguments(
    output_dir="./results_config1",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=5,
    learning_rate=3e-4,
    logging_dir="./logs_config1",
    logging_steps=50,
    save_steps=1000,
    save_total_limit=1,
    do_eval=True,               # ✅ compatible with older versions
    report_to="none"
)

trainer_config1 = Trainer(
    model=model_config1,
    args=training_args_config1,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test
)

trainer_config1.train()
trainer_config1.save_model("./results_config1")
```

```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load tokenizer and fine-tuned model
tokenizer = AutoTokenizer.from_pretrained("t5-small")
config1_model = AutoModelForSeq2SeqLM.from_pretrained("./results_config1/checkpoint-1000")

# Load sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Summary + Sentiment function
def summarize_and_analyze(review_text):
    if not review_text.strip():
        return "Please enter a valid review.", ""

    # Summarization
    input_text = f"summarize: {review_text}"
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = config1_model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_length=60,
        min_length=10,
        num_beams=4,
        early_stopping=True
    )
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Sentiment
    sentiment = sentiment_pipeline(review_text)[0]
    sentiment_text = f"{sentiment['label']} ({round(sentiment['score'] * 100, 2)}%)"

    return summary, sentiment_text

# Sample reviews
example_reviews = [
    "Not great.",
    "Customer service was prompt and helped resolve my issue quickly.",
    "The product arrived two days late and had multiple defects. I contacted customer service and they were rude and unhelpful. Will never buy again.",
    "Amazing product! Exceeded all my expectations and the delivery was super fast.",
    "The item looked cheap and didn’t work as advertised. Waste of money."
]

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🧠 Config1 Fine-Tuned Model - Review Summarizer & Sentiment Analysis")
    gr.Markdown("Enter a review below or select an example to generate both summary and sentiment.")

    with gr.Row():
        dropdown = gr.Dropdown(choices=example_reviews, label="🔽 Example Reviews", interactive=True)
        review_input = gr.Textbox(label="✍️ Your Review", placeholder="Type or paste your review here...")

    dropdown.change(fn=lambda x: gr.update(value=x), inputs=dropdown, outputs=review_input)

    run_button = gr.Button("🚀 Generate Summary & Sentiment")

    with gr.Row():
        summary_out = gr.Textbox(label="🧾 Config1 Summary")
        sentiment_out = gr.Textbox(label="📊 Sentiment")

    run_button.click(fn=summarize_and_analyze, inputs=review_input, outputs=[summary_out, sentiment_out])

demo.launch()
```

```python
from datasets import Dataset
import pandas as pd

# Load your labeled dataset
df = pd.read_csv("ReviewLarge.csv")

# Rename columns for clarity
df = df.rename(columns={"review": "input", "summary": "target"})

# Convert to Hugging Face dataset
dataset = Dataset.from_pandas(df)

# Split into train/test (80/20)
train_test = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = train_test["train"]
test_dataset = train_test["test"]
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("t5-small")

def preprocess_function(examples):
    inputs = tokenizer(examples["input"], truncation=True, padding="max_length", max_length=512)
    targets = tokenizer(examples["target"], truncation=True, padding="max_length", max_length=60)
    inputs["labels"] = targets["input_ids"]
    return inputs

train_data = train_dataset.map(preprocess_function)
test_data = test_dataset.map(preprocess_function)
```

```python
from transformers import TrainingArguments, Trainer, AutoModelForSeq2SeqLM

model_config2 = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

training_args_config2 = TrainingArguments(
    output_dir="./results_config2",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=4,              # ⬅️ Fewer epochs
    learning_rate=3e-5,              # ⬅️ Slightly smaller LR
    logging_dir="./logs_config2",
    logging_steps=10,
    save_steps=500,
    save_total_limit=1,
    report_to="none",
    do_eval=True
)

trainer_config2 = Trainer(
    model=model_config2,
    args=training_args_config2,
    train_dataset=train_data,
    eval_dataset=test_data
)

trainer_config2.train()
trainer_config2.save_model("./results_config2/checkpoint-final")
```
